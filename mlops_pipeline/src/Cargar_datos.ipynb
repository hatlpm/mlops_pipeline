# Carga e Ingesta de Datos
## Proyecto: Modelo Predictivo de Riesgo Crediticio

**Objetivo:** Leer el archivo fuente `Base_de_datos.xlsx`, verificar que 
la carga fue exitosa y exportar los datos como `Base_de_datos.csv` 
en la raíz del proyecto.

**Input:** `../Base_de_datos.xlsx`  
**Output:** `../Base_de_datos.csv`  
**Autor:** Harrison

```python
# ============================================================
# BLOQUE 1: Importación de librerías
# ============================================================

import pandas as pd        # Manipulación de datos tabulares
import os                  # Manejo de rutas y sistema de archivos
```

```python
# ============================================================
# BLOQUE 2: Definición de rutas
# ============================================================

# Ruta relativa al .xlsx desde la ubicación de este notebook (src/)
ruta_xlsx = "../../Base_de_datos.xlsx"

# Verificar que el archivo existe antes de intentar leerlo
if os.path.exists(ruta_xlsx):
    print(f"✓ Archivo encontrado en: {os.path.abspath(ruta_xlsx)}")
else:
    print(f"✗ Archivo NO encontrado en: {os.path.abspath(ruta_xlsx)}")
```

```python
# ============================================================
# BLOQUE 3: Lectura del archivo fuente
# ============================================================

# Leer el .xlsx y cargarlo en un DataFrame
df = pd.read_excel(ruta_xlsx)

print(f"✓ Archivo cargado exitosamente")
print(f"  Filas:    {df.shape[0]}")
print(f"  Columnas: {df.shape[1]}")
```

```python
# ============================================================
# BLOQUE 4: Verificación mínima de la carga
# ============================================================

# Primeras 5 filas para confirmar que el contenido tiene sentido
print("=== PRIMERAS 5 FILAS ===")
display(df.head())

# Tipos de datos por columna
print("\n=== TIPOS DE DATOS ===")
print(df.dtypes)

# Conteo de valores nulos por columna
print("\n=== VALORES NULOS ===")
print(df.isnull().sum())
```

```python
# ============================================================
# BLOQUE 5: Exportación a CSV
# ============================================================

# Ruta de destino del CSV en la raíz del proyecto
ruta_csv = "../../Base_de_datos.csv"

# Exportar sin incluir el índice de pandas como columna extra
df.to_csv(ruta_csv, index=False)

print(f"✓ CSV exportado exitosamente en: {os.path.abspath(ruta_csv)}")
```
